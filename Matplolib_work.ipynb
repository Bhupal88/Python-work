# Matplotlib basic visuals

Here I just work on creating simple figures and testing Matplotlib concepts such as setlimit, insert, subplots, figsize, etc.
***

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes2=fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2=fig.add_axes([0.2,0.5,0.4,0.4])
axes2.plot(x,y)
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom')
```

```python
fig=plt.figure()
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
fig=plt.figure()
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,z)
plt.tight_layout()
```

```python
fig=plt.figure()
fig,axes=plt.subplots(nrows=1,ncols=2, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(x,z)
plt.tight_layout()
```
